```python

headers = {
    "accept": "application/json",
    "Authorization": "Bearer {yourAPIkey} "
}
api_key = '{yourAPIkey}'

def get_top_movies_by_year(year):
    base_url = 'https://api.themoviedb.org/3/discover/movie'
    
    params = { 
        'api_key': api_key,
        'primary_release_year': year,
        'sort_by': 'vote_average.desc',
        'vote_count.gte': 500
    } #search parameters for the query
    
    response = requests.get(base_url, headers=headers, params=params)
    data = response.json()

    movies = []

    for movie in data['results'][:10]:
        movies.append({
            "year": year,
            "title": movie['title'],
            "rating": movie['vote_average'],
            "release_date": movie['release_date'],
            "genre_ids": movie['genre_ids']
                })
    return movies
```

```python
get_top_movies_by_year(2015)
```

```python
import pandas as pd
```

```python
all_movies = []
for year in range(2015,2026):
    movies_by_year = get_top_movies_by_year(year)
    all_movies.extend(movies_by_year) #extend instead of append to put items from one list into another
df = pd.DataFrame(all_movies)
   
```
